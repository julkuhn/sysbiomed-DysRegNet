```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#path to the patient specific network, computed once with the GTEX reference network and once with the TCGA
tcga_breast_path = '/nfs/data2/dysregnet_gtex/results/tcga_breast_no_confounders_binary.fea'
tcga_lung_path = '/nfs/data2/dysregnet_gtex/results/tcga_lung_no_confounders_binary.fea'
tcga_breast_gtex_lung_path = '/nfs/data2/dysregnet_gtex/results/cross_tissue/tcga_breast_gtex_lung.fea'
tcga_lung_gtex_breast_path = '/nfs/data2/dysregnet_gtex/results/cross_tissue/tcga_lung_gtex_breast.fea'

tcga_breast_net = pd.read_feather(tcga_breast_path)
tcga_lung_net = pd.read_feather(tcga_lung_path)
tcga_breast_gtex_lung_net = pd.read_feather(tcga_breast_gtex_lung_path)
tcga_lung_gtex_breast_net =pd.read_feather(tcga_lung_gtex_breast_path)
```

```python
#compare tcga breast and tcga breast and gtex lung
tcga_breast_net = tcga_breast_net.set_index(tcga_breast_net.columns[0])
tcga_breast_net
```

```python
tcga_breast_gtex_lung_net
```

```python
common_columns = tcga_breast_net.columns.intersection(tcga_breast_gtex_lung_net.columns)
common_rows = tcga_breast_net.index.intersection(tcga_breast_gtex_lung_net.index)
```

```python
print(f'shared columns between gtex and tcga breast networks: {len(common_columns)}')
print(f'shared rows between gtex and tcga breast networks {len(common_rows)}')
```

```python
# only keep genes
tcga_breast_common = tcga_breast_net.loc[common_rows, common_columns]
tcga_breast_gtex_lung_common = tcga_breast_gtex_lung_net.loc[common_rows, common_columns]
print(tcga_breast_common.shape)
print(tcga_breast_gtex_lung_common.shape)
```

```python
#make sure rows and column are in the same order
# Ensure rows and columns are in the same order in both DataFrames
tcga_breast_common = tcga_breast_net.loc[common_rows, common_columns].sort_index().sort_index(axis=1)
tcga_breast_gtex_lung_common = tcga_breast_gtex_lung_net.loc[common_rows, common_columns].sort_index().sort_index(axis=1)
```

```python

tcga_breast_common = tcga_breast_common.astype(int)
tcga_breast_gtex_lung_common = tcga_breast_gtex_lung_common.astype(int)
intersection = (tcga_breast_common & tcga_breast_gtex_lung_common).sum(axis=1)
union = ((tcga_breast_common + tcga_breast_gtex_lung_common) > 0).sum(axis=1)

# Store results in a new DataFrame
comparison_df = pd.DataFrame({
    'both_one': intersection,  # Count of 1s in both DataFrames
    'either_one': union,       # Count of 1s in at least one DataFrame
    'jaccard_similarity': intersection / union  # Optional: Jaccard index
})

# Print results
print(comparison_df)
```

```python
#plot the distribution
# Set Seaborn style for better visuals
sns.set_style("whitegrid")

# Plot histogram
plt.figure(figsize=(8, 5))
sns.histplot(comparison_df['jaccard_similarity'], bins=20, kde=True, color="blue")

# Labels and title
plt.xlabel("Jaccard Similarity Score", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.title(f"Jaccard Similarity Scores", fontsize=14)

# Show plot
plt.show()
```
